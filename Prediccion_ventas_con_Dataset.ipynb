```python
#Modelo de Prediccion de ventas con Dataset
#autor: Danny Castilla 26.11.2023
#https://github.com/DCASTILLAV/

#importamos librerias
import pandas as pd
import numpy as np

#import matplotlib as plt
import matplotlib.pyplot as plt

#creamos nuestro dataframe principal y lo almacenamos en la variable df
df=pd.read_csv('https://raw.githubusercontent.com/DCASTILLAV/Datasets/main/ventasSince2017.csv')

#mostramos la cantidad de filas y columnas que hay en nuestro dataframe
#en este caso hay 6 filas y dos columnas
df.shape
```

```python
#mostramos todos los datos de nuestro DataFrame df
df
```

```python
#colocamos los datos de años en la variable x_train
x_train=df.iloc[:,0]
x_train
```

```python
#colocamos los totales de venta en la variable y_train
y_train=df.iloc[:,1]
y_train
```

```python
#aqui solo estamos haciendo una prediccion en base 2
anio =2024

#entrenamos nuestros datos con polinomio de grado 2 con polyfit
coef=np.polyfit(x_train,y_train,2)

#almacenamos la prediccion en la variable p
#como le hemos pasado el parametro del año (anio) y coeficiente (coef)que deseamos calcular eso se almacenará en p
p= np.polyval(coef, anio)

#mostramos datos
#anteponemos la f para indicar a python que le vamos a pasar parametros y de esta forma poder mostrar los valores
#del año y la prediccion
print (f"Para el año {anio} la predicción de Ventas es {p} ------ nos hemos basado en el polinomio de grado 2")
```

```python
#Para mejorar la precicion de las predicciones
#nos apoyaremos en una funcion
#las funciones de declaran anteponiendo la palabra def
#siendo x1

def fx (x1, coef):
    fx = 0
    n = len(coef) - 1
    for p in coef:
        fx = fx + p*x1**n
        n = n - 1
    return fx
```

```python
anio = 2024
for i in range(0,10):
    coef = np.polyfit(x_train,y_train,i)
    p = np.polyval(coef, anio)

    print(f"Para grado {i} la predicción de ventas para el año: {anio} es {p}")
    x1 = np.linspace(2017, anio + 1, 1000)
    y1 = fx(x1, coef) # funcion

    plt.figure(figsize=[20,10])

    plt.title("Pronóstico de ventas vs año. para el grado: " + str(i))

    plt.scatter(x_train,y_train,s=120,c='blueviolet')
    plt.plot(x1,y1,"--",linewidth=3,color='orange')
    plt.scatter(anio,p,s=200,c='red')
    plt.yticks(range(100,320,20))
    plt.grid("on")
    ax=plt.gca()
    ax.set_xlabel("$años$")
    ax.set_ylabel("$Total ventas$")
    #plt.savefig("img" + str(i)+".jpg", dpi=600)
    plt.show()
```

```python
#detectamos el modelo de grado que mas eficiente sea
# y sea la prediccion que sea mas aceptable o se asemeje a la realidad
#para ello utilizamos el error de medio cuadratico (MSE)

#Error Cuadrático Medio (MSE o MES) Es una de las funciones de pérdida de regresión más comunes.
#En el error cuadrático medio, también conocido como pérdida L2,
#calculamos el error elevando al cuadrado la diferencia entre el valor predicho
# y el valor real y promediando el conjunto de datos.

#midiendo el error de medio cuadratico podemos validar que grado conviene mas a nuestro modelo
#la grafica nos muestra que mientras mas grado sea es menos optimo
#aqui debemos elegir la estimas mas optima y que tenga sentido
#la predicció mas optima que hemos elegido es la predicion n°11 ya que no es ni tan baja ni tan alta y
#se mantiene en un termino medio y realista
#------ CON ESTA DATA podemos hacer ya solo nuestro modelo en base a ese analisis---
#------ solo consideraremos la proyeccion en grado 11 ----------


anio = 2024
grado = np.arange(0, 30 + 1, 1) #de 0 a 100 de 1 en 1
aproxi = np.array([])
y_pred_vec = np.array([])
for i in grado:
    coef = np.polyfit(x_train,y_train,i)
    p = np.polyval(coef, anio)
    aproxi = np.append(aproxi, p)

    # para la MSE
    y_pred_vec = np.array([])
    for j in x_train: # para la MSE
        y_pred = np.polyval(coef, j)
        y_pred_vec = np.append(y_pred_vec,y_pred)
    #print(f"los y: {y}")
    #print(f"los y_aprox: {y_pred_vec}")
    MSE = (sum( (y_train - y_pred_vec)**2 ))/len(y_train)
    print(f"Para grado {i} El MES es: {MSE}")
plt.figure(figsize=[20,10])
plt.title("Grado del polinomio vs predicción")
plt.plot(grado,aproxi,"--",linewidth=3,color='red')
plt.grid("on")
plt.show()
```
